# Using the above defined Unigram to generate possible candidate words for each error type 

```python
alphabets = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 
             'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
```

```python
def insertion_candidates(word):
    cand_words = []
    for i in alphabets:
        string = ""
        for j in (range(len(word) + 1)):
            if (j == len(word)):
                string = word[:(j + 1)] + i + word[(j + 1):]
            else:   
                string = word[:j] + i + word[j:]
            if (string in unigram.keys()):
                cand_words.append(string)
    return cand_words
```

```python
def deletion_candidates(word):
    cand_words = []
    for i in (range(len(word))):
        string = word.replace(word[i], "")
        if (string in unigram.keys()):
                cand_words.append(string)
    return cand_words
```

```python
def substitution_candidates(word):
    cand_words = []
    for i in alphabets:
        for j in (range(len(word))):
            string = word.replace(word[j], i)
            if (string in unigram.keys()):
                cand_words.append(string)
    return cand_words
```

```python
def swap(text, ch1, ch2):
    text = text.replace(ch2, '!',)
    text = text.replace(ch1, ch2)
    text = text.replace('!', ch1)
    return text
```

```python
def transposition_candidates(word):
    cand_words = []
    for i in (range(len(word))):
        if (i < len(word) - 1):
            string = swap(word, word[i], word[i + 1])
        if (string in unigram.keys()):
                cand_words.append(string)
    return cand_words
```

```python
def create_candidate_words(word):
    candidates = []
    candidates += deletion_candidates(word)
    candidates += substitution_candidates(word)
    candidates += transposition_candidates(word)
    candidates += insertion_candidates(word)
    return candidates
```

```python
def correct_word_checker(word): #Function to check whether or not the word is already in its correct state
    if (word not in unigram.keys()):
        return False
    else:
        return True
```

# Identifying the error type in each candidate word and then using the above defined "Error Model Edit" tables to calculate the occurrence probability of each error. This function returns a dictionary of candidate words along side their probabilities. The candidate word with the highest probability is considered the correct spelled version of the input word.

```python
def candidate_words_probabilities(word):
    candidates = create_candidate_words(word)
    candidates_probability = {}
    for i in candidates:
        candidates_probability[i] = 0
        
    #THE CHARACTER BEFORE THE DELETED ONE - THE CHARACTER THAT WAS DELETED - COUNT
    DELETED = []
    DELETED_CAND = []
    for i in candidates:
        if (len(word) > len(i)): #deletion
            deleted_char = ""
            deleted_idx = 0
            for j in range(len(word)):
                if (j >= len(i)):
                    deleted_char = word[j]
                    deleted_idx = j
                    if (deleted_idx == 0):
                        DELETED.append("@" + ', ' + deleted_char)
                        DELETED_CAND.append(i)
                    else:
                        DELETED.append(word[j - 1] + ', ' + deleted_char)
                        DELETED_CAND.append(i)
                    break
                elif (word[j] != i[j]):
                    deleted_char = word[j]
                    deleted_idx = j
                    if (deleted_idx == 0):
                        DELETED.append("@" + ', ' + deleted_char)
                        DELETED_CAND.append(i)
                    else:
                        DELETED.append(word[j - 1] + ', ' + deleted_char)
                        DELETED_CAND.append(i)
                    break
    index = 0
    index2 = 0
    for i in deletion.keys():
        for j in DELETED:
            if (i == j):
                candidates_probability[DELETED_CAND[index2]] = list(deletion.values())[index]/sum(deletion.values())
                index2 = index2 + 1
        index = index + 1
        
        
    #THE CHARACTER BEFORE THE INSERTED ONE - THE CHARACTER THAT WAS INSERTED - COUNT
    INSERTED = []
    INSERTED_CAND = []
    for i in candidates:
        if (len(word) < len(i)): #insertion
            for j in range(len(i)):
                if (j < len(word)):
                    if (word[j] != i[j]):
                        inserted_char = i[j]
                        if (j == 0):
                            INSERTED.append("@" + ', ' + inserted_char)
                            INSERTED_CAND.append(i)
                        else:
                            INSERTED.append(word[j - 1] + ', ' + inserted_char)
                            INSERTED_CAND.append(i)
                        break
                else:
                    inserted_char = i[j]
                    if (j == 0):
                        INSERTED.append("@" + ', ' + inserted_char)
                        INSERTED_CAND.append(i)
                    else:
                        INSERTED.append(word[j - 1] + ', ' + inserted_char)
                        INSERTED_CAND.append(i)
                        
    index = 0
    index2 = 0
    for i in insertion.keys():
        for j in INSERTED:
            if (i == j):
                candidates_probability[INSERTED_CAND[index2]] = list(insertion.values())[index]/sum(insertion.values())
                index2 = index2 + 1
        index = index + 1
        
    #THE CORRECT LETTER - THE WRONGLY USED ONE - COUNT
    SUBSTITUTED = []
    SUBSTITUTED_CAND = []
    for i in candidates:
        if (len(word) == len(i)): #substitution
            for j in range(len(i)):
                if (word[j] != i[j]):
                    substituted_char = i[j]
                    SUBSTITUTED.append(word[j] + ', ' + substituted_char)
                    SUBSTITUTED_CAND.append(i)
                    
    index = 0
    index2 = 0
    for i in substitution.keys():
        for j in SUBSTITUTED:
            if (i == j):
                candidates_probability[SUBSTITUTED_CAND[index2]] = list(substitution.values())[index]/sum(substitution.values())
                index2 = index2 + 1
        index = index + 1
        
    #THE FIRST CHARACTER THAT NEEDS TO BE SWAPPED - THE SECOND CHARACTER THAT NEEDS TO BE SWAPPED - COUNT
    TRANSPOSED = []
    TRANSPOSED_CAND = []
    for i in candidates:
        if (len(i) == len(word)): #transposition
            if (sorted(i) == sorted(word)):
                for idx, j in enumerate(i):
                    if (idx < len(i) - 1):
                        if (i[idx] == word[idx + 1]):
                            TRANSPOSED.append(i[idx] + ', ' + word[idx])
                            TRANSPOSED_CAND.append(i)
    index = 0
    index2 = 0
    for i in transposition.keys():
        for j in TRANSPOSED:
            if (i == j):
                candidates_probability[TRANSPOSED_CAND[index2]] = list(transposition.values())[index]/sum(transposition.values())
                index2 = index2 + 1
        index = index + 1

    return candidates_probability
```
